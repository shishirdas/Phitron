Max and Min problem

```python
inp = input()

numbers = inp.split()

x = int(numbers[0])
y = int(numbers[1])
z = int(numbers[2])

min = x 
max = x 

## min
if y < min :
    min = y 
if z < min :
    min = z 


## max 

if y > max :
    max = y 

if z > max :
    max = z 


print(min , max)
```

C. Even, Odd, Positive and Negative

```python
n = int(input()) 

inp = input()

numbers= inp.split()

positive =0 ; negative =0 ; even =0 ; odd =0 


for i in range(n):
    x = int(numbers[i])
    # positive or negative
    if x > 0 :
        positive+=1 
    elif x < 0 :
        negative+=1 
    
    # odd or even
    if x % 2 !=0 :
        odd+=1 
    else:
        even+=1 


print("Even:",even)
print("Odd:",odd)
print("Positive:",positive)
print("Negative:",negative)
```

#### Digits

```python
t = int(input())

for i in range(t):
    number = int(input())
    if number ==0 :
        print(0)
        continue
    while number > 0 :
        print(number%10, end=" ")
        number//=10
    print()
```
